```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1 = np.array(
[0.0007569581235367584,0.010019962360126202,0.012464057871583872,0.015423050077991112,0.01847674981712565])
w2 = np.array([0.00867384, 0.01762371, 0.00885913, 0.01549003, 0.01450691])
w3 = np.array([0.00107091, -0.00135008, -0.00088579,  0.00052553,
        0.0015077 ])
w4 = np.array([0.0076706, 0.00769535,0.00431324,0.00860999,0.00643872])
```

```python
inn = 100000000
als = []
w11 = 100000000*(1+w1)
w22 = w11[-1] * (1+w2)
w33 = w22[-1] * (1+w3)
w44 = w33[-1] * (1+w4)
a = np.append(w11, [w22,w33,w44])
plt.rcParams['figure.figsize'] = (8.0, 4.0)
plt.plot(a)
```

```python
from matplotlib.pyplot import MultipleLocator
x_values=list(range(1,21))
y_values=a/inn

plt.rcParams['figure.figsize'] = (14.0, 4.0)
plt.title('Portfolio Unit Value from 03-08 to 04-02',fontsize=14)
plt.tick_params(axis='both',which='major',labelsize=14)
plt.xlabel('Trading Dates',fontsize=14)
plt.ylabel('Accumulative Return',fontsize=14)
x_major_locator=MultipleLocator(1)
#把y轴的刻度间隔设置为10，并存在变量里
ax=plt.gca()
#ax为两条坐标轴的实例
ax.xaxis.set_major_locator(x_major_locator)
plt.plot(x_values,y_values)
```

```python
rets = a/inn
print(rets)
sharpe = (np.mean(rets) - 0.03/252) /  np.std(rets)
print(sharpe)
print(np.std(rets))
import pandas as pd
pd.Series(rets).to_csv('discretionary.csv')
```
